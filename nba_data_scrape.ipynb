```python
years = list(range(2014, 2024))
```

```python
url_start = 'https://www.basketball-reference.com/leagues/NBA_{year}.html'
```

```python
requests_per_minute_limit = 20
seconds_per_minute = 60

request_counter = 0

for year in years:
    if request_counter >= requests_per_minute_limit:
        print("Reached the request limit. Resting for 1 minute...")
        time.sleep(65)
        request_counter = 0  # Reset the request counter after the rest
            
    url = url_start.format(year=year)
    data = requests.get(url)

    with open(f'adv_team/adv_{year}.html', 'w+') as f:
        f.write(data.text)
            
    request_counter += 1

print("All requests completed.")
```

```python
dfs = []

for year in years:
    with open(f'adv_team/adv_{year}.html') as f:
        soup = BeautifulSoup(f, 'html.parser')
    soup.find('tr', class_="over_header").decompose()
    adv_table = soup.find('table', id='advanced-team')
    adv = pd.read_html(str(adv_table))[0]
    adv['Year'] = year

    dfs.append(adv)
```

```python
advs = pd.concat(dfs)
```
